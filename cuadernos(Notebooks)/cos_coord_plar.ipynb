# Anlisis de la funcin Coseno en coordenadas Polares  

##  Introduccin  
En este experimento interactivo, analizaremos cmo cambia la funcin **coseno** en coordenadas polares al modificar su **frecuencia angular**. Esta funcin es fundamental en oscilaciones y anlisis de seales.  

Exploraremos las siguientes transformaciones:  
- **cos()**  Representa la oscilacin estndar.  
- **cos(2)**  Duplica la frecuencia, aumentando la cantidad de ciclos.  
- **cos(/2)**  Reduce la frecuencia, generando curvas ms amplias.  

Para visualizar estos cambios, utilizaremos **grficos en coordenadas polares** con interactividad.  

## Conceptos  
1. **Coordenadas Polares**: Sistema en el cual los puntos se representan mediante un radio \( r \) y un ngulo $( \theta )$.  
2. **Frecuencia Angular**: Un factor multiplicativo en el argumento de `cos()`, que afecta la cantidad de oscilaciones en un ciclo.  
3. **Interactividad**: Uso de `ipywidgets` para manipular dinmicamente la frecuencia y observar sus efectos en tiempo real.  


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

#  Funcin para graficar las curvas coseno en coordenadas polares
def graficar_cos(frecuencia):
    theta = np.linspace(0, 2 * np.pi, 500)  # Valores de  de 0 a 2
    r = np.cos(frecuencia * theta)  # Funcin coseno en coordenadas polares

    #  Crear la figura y eje polar
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(6, 6))
    ax.plot(theta, r, color='cyan', linewidth=2, label=f'cos({frecuencia})')

    #  Personalizar el grfico
    ax.set_facecolor('black')  # Fondo oscuro
    ax.spines['polar'].set_color('white')  # Borde blanco
    ax.grid(color='gray', linestyle='dotted')  # Cuadrcula gris
    ax.set_title(f"Curva Polar de cos({frecuencia})", color='white', fontsize=14)
    ax.legend(loc='upper right', fontsize=12, facecolor='black', edgecolor='cyan')

    #  Mostrar la grfica
    plt.show()

#  Widget interactivo
frecuencia_slider = widgets.FloatSlider(
    min=0.5, max=4, step=0.5, value=1, description="Frecuencia", style={'description_width': 'initial'}
)

#  Conectar el widget con la funcin
widgets.interactive(graficar_cos, frecuencia=frecuencia_slider)

```

```python
import numpy as np
import plotly.graph_objects as go

# Definir el rango de valores de  (de 0 a 2)
theta = np.linspace(0, 2 * np.pi, 200)

# Crear los frames para la animacin con diferentes frecuencias
frames = []
for f in np.linspace(0.5, 4, 50):  # Variamos la frecuencia de 0.5 a 4
    r = np.cos(f * theta)  # Funcin en coordenadas polares
    frame = go.Frame(data=[go.Scatterpolar(r=r, theta=np.degrees(theta), mode="lines", line=dict(color="cyan", width=2))])
    frames.append(frame)

# Configuracin de la figura principal
fig = go.Figure(
    data=[go.Scatterpolar(r=np.cos(theta), theta=np.degrees(theta), mode="lines", line=dict(color="cyan", width=2))],
    layout=go.Layout(
        title=" Animacin interactiva de la funcin Coseno en coordenadas Polares",
        polar=dict(bgcolor="black", radialaxis=dict(showticklabels=False, showgrid=True)),
        plot_bgcolor="black",
        paper_bgcolor="black",
        font=dict(color="white"),
        updatemenus=[{
            "buttons": [
                {"args": [None, {"frame": {"duration": 100, "redraw": True}, "fromcurrent": True}], "label": " Iniciar", "method": "animate"},
                {"args": [[None], {"frame": {"duration": 0, "redraw": True}, "mode": "immediate"}], "label": " Detener", "method": "animate"},
            ],
            "direction": "left",
            "pad": {"r": 10, "t": 10},
            "showactive": True,
            "type": "buttons",
            "x": 0.1,
            "xanchor": "right",
            "y": 1.1,
            "yanchor": "top"
        }]
    ),
    frames=frames
)

# Mostrar el grfico interactivo
fig.show()
```

## Resultados y conclusiones  
- A medida que aumenta la frecuencia, el nmero de ciclos en la grfica se incrementa.  
- Al disminuir la frecuencia, las oscilaciones se hacen ms amplias y suaves.  
- Este tipo de representacin es til en el anlisis de seales, transformadas de Fourier y fenmenos peridicos en la naturaleza.
