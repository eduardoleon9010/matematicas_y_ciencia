## üìö **Fundamentos matem√°ticos**
La serie de Fourier de una funci√≥n peri√≥dica $( f(x) )$ se expresa como:

$$
f(x) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n\omega x) + b_n \sin(n\omega x) \right)
$$

Para la **onda cuadrada**, la aproximaci√≥n con \( N \) t√©rminos est√° dada por:

$$
f_N(x) = \sum_{k=1,3,5,\dots}^{N} \frac{4}{k\pi} \sin(kx)
$$

donde solo aparecen los t√©rminos impares. A medida que aumentamos \( N \), la aproximaci√≥n mejora.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from ipywidgets import interact, IntSlider
from IPython.display import HTML

# Funci√≥n para calcular la serie de Fourier de una funci√≥n peri√≥dica
def fourier_series(x, n_terms):
    """Aproxima una funci√≥n peri√≥dica con una serie de Fourier."""
    y = np.zeros_like(x)
    for n in range(1, n_terms + 1, 2):  # Solo t√©rminos impares para onda cuadrada
        y += (4 / (n * np.pi)) * np.sin(n * x)
    return y

# Crear figura para la animaci√≥n
fig, ax = plt.subplots(figsize=(8, 4))
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title("Aproximaci√≥n de una Onda Cuadrada con Serie de Fourier")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid()

x = np.linspace(0, 2 * np.pi, 400)
line, = ax.plot(x, fourier_series(x, 1), 'b', label="Serie de Fourier")
ax.legend()

# Funci√≥n de actualizaci√≥n de la animaci√≥n
def update(n):
    line.set_ydata(fourier_series(x, n))
    return line,

# Crear animaci√≥n
ani = animation.FuncAnimation(fig, update, frames=np.arange(1, 20, 2), interval=500, blit=True)
HTML(ani.to_jshtml())

# Agregar interactividad con sliders
def interactive_fourier(n_terms):
    plt.figure(figsize=(8, 4))
    plt.plot(x, fourier_series(x, n_terms), 'b', label=f"{n_terms} t√©rminos")
    plt.title("Serie de Fourier Interactiva")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.legend()
    plt.show()

interact(interactive_fourier, n_terms=IntSlider(min=1, max=50, step=1, value=5));
```

## **Resultados**
Tras el desarrollo y ejecucion del script interactivo y observamos:

-> C√≥mo la aproximaci√≥n mejora al agregar m√°s t√©rminos.

-> El **efecto de la oscilaci√≥n de Gibbs**, que aparece cerca de los saltos de la onda cuadrada.

-> C√≥mo la **Serie de Fourier** nos ayuda a representar funciones complicadas con solo senos y cosenos.
