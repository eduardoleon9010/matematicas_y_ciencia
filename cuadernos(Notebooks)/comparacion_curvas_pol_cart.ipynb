#  **Comparacin de curvas Polares y Cartesianas**  

##  **Introduccin**  
Las curvas pueden representarse en **coordenadas cartesianas** $((x, y))$ o en **coordenadas polares** $((r, \theta))$. Dependiendo del sistema utilizado, una misma figura puede tomar diferentes expresiones matemticas.  

En este cuaderno, exploraremos cmo una misma curva se visualiza en ambos sistemas de coordenadas mediante ejemplos dinmicos.  

###  **Objetivos de aprendizaje**  
 Comparar la representacin de curvas en coordenadas **polares** y **cartesianas**  
 Observar la evolucin de cada curva mientras un punto se mueve sobre su trayectoria  
 Implementar visualizaciones interactivas en **Python** con `matplotlib` y `numpy`  



### **Coordenadas Cartesianas**  
El sistema cartesiano representa puntos con dos valores:  

$$
(x, y)
$$

Donde \( x \) y \( y \) son las posiciones en los ejes horizontal y vertical, respectivamente.  

### **Coordenadas Polares**  
El sistema polar representa los puntos con un **radio** \( r \) y un **ngulo** \( \theta \):  

$$
r = f(\theta)
$$

Donde:  
 \( r \) es la distancia desde el origen  
 $( \theta )$ es el ngulo medido desde el eje horizontal  

Se convierte a coordenadas cartesianas con:  

$$
x = r \cos(\theta), \quad y = r \sin(\theta)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import numpy as np
import plotly.graph_objects as go
```

```python
# Definir funcin para graficar curvas en coordenadas cartesianas y polares
def graficar_curvas():
    t = np.linspace(0, 2 * np.pi, 500)

    # Crculo en coordenadas polares y cartesianas
    r_circulo = np.ones_like(t)  # Radio constante
    x_circulo = np.cos(t)
    y_circulo = np.sin(t)

    # Cuadrado en coordenadas polares (aproximado)
    x_cuadrado = np.sign(np.cos(t))
    y_cuadrado = np.sign(np.sin(t))
    r_cuadrado = np.maximum(abs(x_cuadrado), abs(y_cuadrado))

    # Configuracin de grficos con ejes polares y cartesianas correctamente asignados
    fig = plt.figure(figsize=(10, 8))

    ax1 = fig.add_subplot(221)  # Eje cartesiano
    ax2 = fig.add_subplot(222, polar=True)  # Eje polar
    ax3 = fig.add_subplot(223)  # Eje cartesiano
    ax4 = fig.add_subplot(224, polar=True)  # Eje polar

    # Grfica del crculo
    ax1.plot(x_circulo, y_circulo, 'b')
    ax1.set_title("Crculo en Coordenadas Cartesianas")
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.grid()

    ax2.plot(t, r_circulo, 'b')
    ax2.set_title("Crculo en Coordenadas Polares")

    # Grfica del cuadrado
    ax3.plot(x_cuadrado, y_cuadrado, 'r')
    ax3.set_title("Cuadrado en Coordenadas Cartesianas")
    ax3.set_xlabel("X")
    ax3.set_ylabel("Y")
    ax3.grid()

    ax4.plot(t, r_cuadrado, 'r')
    ax4.set_title("Cuadrado en Coordenadas Polares")

    # Mostrar grficos
    plt.tight_layout()
    plt.show()

# Llamar a la funcin
graficar_curvas()
```

```python
# Funcin para graficar curvas con valores dinmicos
def graficar_curvas(escalar):
    t = np.linspace(0, 2 * np.pi, 500)

    # Crculo (ajustable con el escalar)
    x_circulo = escalar * np.cos(t)
    y_circulo = escalar * np.sin(t)

    # Cuadrado en coordenadas cartesianas (ajustado)
    x_cuadrado = escalar * np.sign(np.cos(t))
    y_cuadrado = escalar * np.sign(np.sin(t))

    # Configuracin de grficos
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Grfica del crculo
    axs[0].plot(x_circulo, y_circulo, 'b')
    axs[0].set_title("Crculo en Coordenadas Cartesianas")
    axs[0].set_xlabel("X")
    axs[0].set_ylabel("Y")
    axs[0].grid()

    # Grfica del cuadrado
    axs[1].plot(x_cuadrado, y_cuadrado, 'r')
    axs[1].set_title("Cuadrado en Coordenadas Cartesianas")
    axs[1].set_xlabel("X")
    axs[1].set_ylabel("Y")
    axs[1].grid()

    plt.show()

# Crear slider interactivo
slider_escalar = widgets.FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1, description="Escalar:")
widgets.interactive(graficar_curvas, escalar=slider_escalar)
```

```python
# Generar datos para el crculo y el cuadrado
t = np.linspace(0, 2 * np.pi, 500)
x_circulo = np.cos(t)
y_circulo = np.sin(t)
z_circulo = np.sin(3 * t)  # Para darle profundidad en 3D

x_cuadrado = np.sign(np.cos(t))
y_cuadrado = np.sign(np.sin(t))
z_cuadrado = np.cos(3 * t)

# Crear figura 3D con Plotly
fig = go.Figure()

# Crculo en 3D
fig.add_trace(go.Scatter3d(
    x=x_circulo, y=y_circulo, z=z_circulo,
    mode="lines", line=dict(color="blue", width=4),
    name="Crculo"
))

# Cuadrado en 3D
fig.add_trace(go.Scatter3d(
    x=x_cuadrado, y=y_cuadrado, z=z_cuadrado,
    mode="lines", line=dict(color="red", width=4),
    name="Cuadrado"
))

# Configuracin del grfico
fig.update_layout(
    title="Comparacin de curvas en 3D",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=700, height=500
)

# Mostrar grfico interactivo
fig.show()
```
