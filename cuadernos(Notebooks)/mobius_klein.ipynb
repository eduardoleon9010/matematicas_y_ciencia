## 游닀 Introducci칩n

La **Cinta de M칬bius** y la **Botella de Klein** son superficies no orientables fundamentales en topolog칤a.

游늷 **Cinta de M칬bius:**
- Descubierta en 1858 por August Ferdinand M칬bius y Johann Benedict Listing.
- Se forma tomando una cinta rectangular, girando uno de sus extremos 180춿 y peg치ndolo con el otro extremo.
- Tiene solo **una cara y un solo borde**.
- Se aplica en **ingenier칤a, f칤sica y ciencias computacionales**.


游늷 **Botella de Klein:**
- Descubierta en 1882 por Felix Klein.
- Es una superficie cerrada sin borde que **no puede ser embebida en 3D sin intersecciones**.
- Es la extensi칩n de la Cinta de M칬bius en cuatro dimensiones.
- Es utilizada en **geometr칤a, topolog칤a y f칤sica te칩rica**

En este notebook, vamos a **visualizar ambas superficies** utilizando **Python y Plotly**.


```python
#  Importamos las bibliotecas necesarias
import numpy as np
import plotly.graph_objects as go

# Generamos los puntos para la Cinta de M칬bius
def generar_cinta_mobius(n=100, vueltas=1):
    """Genera los puntos 3D de una Cinta de M칬bius con `n` divisiones y `vueltas` completas."""
    u = np.linspace(0, 2 * np.pi * vueltas, n)  # Par치metro que recorre la cinta
    v = np.linspace(-1, 1, 10)  # Par치metro para el ancho de la cinta
    u, v = np.meshgrid(u, v)

    # Ecuaciones param칠tricas de la Cinta de M칬bius
    x = (1 + 0.5 * v * np.cos(u / 2)) * np.cos(u)
    y = (1 + 0.5 * v * np.cos(u / 2)) * np.sin(u)
    z = 0.5 * v * np.sin(u / 2)

    return x, y, z

# Generamos la superficie
x, y, z = generar_cinta_mobius()

# Creamos la visualizaci칩n interactiva con Plotly
fig = go.Figure()
fig.add_trace(go.Surface(x=x, y=y, z=z, colorscale='Viridis'))

# Configuraciones del gr치fico
fig.update_layout(
    title="游늷 Cinta de M칬bius - Representaci칩n en 3D",
    scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Mostramos la gr치fica interactiva
fig.show()
```

```python
# Generamos los puntos para la Botella de Klein
def generar_botella_klein(n=100):
    """Genera los puntos 3D de una Botella de Klein con `n` divisiones."""
    u = np.linspace(0, 2 * np.pi, n)
    v = np.linspace(0, 2 * np.pi, n)
    u, v = np.meshgrid(u, v)

    # Ecuaciones param칠tricas de la Botella de Klein
    x = (2 + np.cos(u / 2) * np.sin(v) - np.sin(u / 2) * np.sin(2 * v)) * np.cos(u)
    y = (2 + np.cos(u / 2) * np.sin(v) - np.sin(u / 2) * np.sin(2 * v)) * np.sin(u)
    z = np.sin(u / 2) * np.sin(v) + np.cos(u / 2) * np.sin(2 * v)

    return x, y, z
# Generamos la superficie de la Botella de Klein
x, y, z = generar_botella_klein()

# Creamos la visualizaci칩n interactiva con Plotly para la Botella de Klein
fig_klein = go.Figure()
fig_klein.add_trace(go.Surface(x=x, y=y, z=z, colorscale='Plasma'))
fig_klein.update_layout(title=" Botella de Klein - Representaci칩n en 3D",
                        scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
                        margin=dict(l=0, r=0, b=0, t=40))
fig_klein.show()
```
