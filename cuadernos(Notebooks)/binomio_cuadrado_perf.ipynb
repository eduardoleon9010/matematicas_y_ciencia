En este notebook explicaremos el concepto del **binomio cuadrado perfecto** y su representaciÃ³n geomÃ©trica.

### Â¿QuÃ© es el binomio cuadrado perfecto?
Es una identidad algebraica de la forma:

$$
(a + b)^2 = a^2 + 2ab + b^2
$$

Nos dice que el cuadrado de la suma de dos tÃ©rminos es igual a la suma de los cuadrados individuales mÃ¡s el doble del producto de ambos.

En este notebook veremos su demostraciÃ³n algebraica y una visualizaciÃ³n geomÃ©trica con Python. ðŸš€

```python
import numpy as np
import plotly.graph_objects as go

# FunciÃ³n para representar grÃ¡ficamente el binomio cuadrado perfecto en 3D
def binomio_cuadrado_3d(a, b):
    X = np.linspace(-a - b, a + b, 50)
    Y = np.linspace(-a - b, a + b, 50)
    X, Y = np.meshgrid(X, Y)
    Z = X**2 + 2*X*Y + Y**2  # ExpansiÃ³n del binomio cuadrado perfecto

    # Crear la figura interactiva con Plotly
    fig = go.Figure()
    fig.add_trace(go.Surface(z=Z, x=X, y=Y, colorscale='viridis'))

    fig.update_layout(
        title=f'RepresentaciÃ³n 3D del Binomio Cuadrado Perfecto: ({a} + {b})Â²',
        scene=dict(
            xaxis_title='Valor de X',
            yaxis_title='Valor de Y',
            zaxis_title='XÂ² + 2XY + YÂ²'
        )
    )
    fig.show()

# ParÃ¡metros iniciales
a = 2
b = 2

# Llamar a la funciÃ³n para visualizar el grÃ¡fico 3D
binomio_cuadrado_3d(a, b)
```
