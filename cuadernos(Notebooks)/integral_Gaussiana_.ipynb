Su soluci贸n exacta est谩 relacionada con la **ra铆z cuadrada de **:

Esta integral es clave en muchas 谩reas de la matem谩tica aplicada, incluyendo la **teor铆a de probabilidades**, donde se usa para derivar la **funci贸n de densidad de la distribuci贸n normal**.

```python
import plotly.graph_objects as go
import numpy as np

# Datos base
num_frames = 50
x = np.linspace(0, 10, 100)
frames = []

# Generar frames para la animaci贸n
for i in range(num_frames):
    y = np.sin(x + (i / 5))  # Variaci贸n en cada fotograma
    frames.append(go.Frame(data=[go.Scatter(x=x, y=y, mode="lines", line=dict(color="blue"))]))

# Figura base
fig = go.Figure(
    data=[go.Scatter(x=x, y=np.sin(x), mode="lines", line=dict(color="blue"))],
    layout=go.Layout(
        title="Animaci贸n de una onda sinusoidal ",
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        updatemenus=[{
            "buttons": [
                {
                    "args": [None, {"frame": {"duration": 50, "redraw": True}, "fromcurrent": True}],
                    "label": "讹 Play",
                    "method": "animate"
                },
                {
                    "args": [[None], {"frame": {"duration": 0}, "mode": "immediate", "redraw": True}],
                    "label": "革 Pause",
                    "method": "animate"
                }
            ],
            "direction": "left",
            "pad": {"r": 10, "t": 87},
            "showactive": False,
            "type": "buttons",
            "x": 0.1,
            "xanchor": "right",
            "y": 0,
            "yanchor": "top"
        }],
        sliders=[{
            "steps": [
                {
                    "args": [[f.name], {"frame": {"duration": 50, "redraw": True}, "mode": "immediate"}],
                    "label": str(i),
                    "method": "animate"
                } for i, f in enumerate(frames)
            ],
            "currentvalue": {"prefix": "Frame: ", "font": {"size": 16}},
        }]
    ),
    frames=frames
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

# Habilitar el renderizado en Colab
pio.renderers.default = "colab"

# Definir la funci贸n de la Integral Gaussiana
def gaussian_integral(a):
    x = np.linspace(-5, 5, 400)
    y = np.exp(-a * x**2)
    integral_values = np.cumsum(y) * (x[1] - x[0])  # Integral acumulada

    fig = go.Figure()

    # Graficar la funci贸n Gaussiana
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Funci贸n Gaussiana', line=dict(color="blue")))

    # Graficar la integral acumulada
    fig.add_trace(go.Scatter(x=x, y=integral_values, mode='lines', fill='tozeroy', name='Integral Acumulada', line=dict(color="orange")))

    # Configurar dise帽o
    fig.update_layout(
        title=f'Integral Gaussiana con a = {a:.2f}',
        xaxis_title='x',
        yaxis_title='f(x)',
        template='plotly_dark',
        width=800,
        height=500
    )

    # Mostrar el gr谩fico directamente en Google Colab
    fig.show()

# Llamar a la funci贸n con un valor de prueba
gaussian_integral(1)
```
