# Curvas polares: Explorando una funcin trascendente en coordenadas polares

##  Introduccin  
Las curvas polares son representaciones matemticas en las que la posicin de un punto en el plano se describe mediante un radio \( r \) y un ngulo \( \theta \). Estas curvas han sido ampliamente estudiadas en matemticas y fsica debido a sus aplicaciones en mecnica orbital, diseo de antenas y modelado de fenmenos naturales (Stewart, 2020).
La **curva mariposa** es un ejemplo fascinante de una funcin trascendente representada en coordenadas polares. Su ecuacin es:

$$
r = e^{\cos(\theta)} - 2\cos(4\theta) + \sin^5\left(\frac{\theta}{12}\right)
$$

Donde ( r ) es la distancia radial y $( \theta )$ es el ngulo en radianes.  
Este tipo de funcin no puede expresarse mediante polinomios finitos y da lugar a patrones geomtricos sorprendentes. La grfica resultante tiene una forma simtrica con ondulaciones que recuerdan a las alas de una mariposa, de ah su nombre.

Adicional, es posible compararla con otras curvas polares notables, como la **Rosa Matemtica** y la **Cardioide**, que tienen aplicaciones geomtricas y en la teora de seales (Weisstein, 2021).  

En este trabajo, exploramos estas curvas de manera interactiva, permitiendo modificar sus ecuaciones y parmetros para comprender mejor su comportamiento.  



```python
import numpy as np
import plotly.graph_objects as go

# Definimos el rango de valores de t
t = np.linspace(0, 12 * np.pi, 1000)

# Definimos la ecuacin de la curva mariposa en coordenadas polares
r = np.exp(np.cos(t)) - 2 * np.cos(4*t) + np.sin(t/12)**5

# Convertimos coordenadas polares a cartesianas
x = r * np.cos(t)
y = r * np.sin(t)

# Creamos la figura interactiva
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Curva Mariposa'))

# Configuracin del grfico
fig.update_layout(
    title="Curva mariposa en coordenadas polares",
    xaxis_title="Eje X",
    yaxis_title="Eje Y",
    showlegend=True,
    template="plotly_dark"
)

# Mostramos la grfica interactiva
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Definimos las ecuaciones en coordenadas polares
def butterfly_curve(theta, a=1, b=2, c=12):
    """Ecuacin de la curva mariposa en coordenadas polares."""
    return np.exp(np.cos(theta)) - b * np.cos(4 * theta) + np.sin(theta / c) ** 5

def rose_curve(theta, k=5):
    """Ecuacin de la rosa matemtica en coordenadas polares."""
    return np.cos(k * theta)

def cardioid_curve(theta, a=1):
    """Ecuacin de la cardioide en coordenadas polares."""
    return a * (1 + np.cos(theta))

# Funcin para graficar
def plot_polar_curve(curve_type="butterfly", a=1, b=2, c=12, k=5):
    theta = np.linspace(0, 2 * np.pi, 1000)

    if curve_type == "butterfly":
        r = butterfly_curve(theta, a, b, c)
    elif curve_type == "rose":
        r = rose_curve(theta, k)
    elif curve_type == "cardioid":
        r = cardioid_curve(theta, a)

    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(6, 6))
    ax.plot(theta, r, label=curve_type.capitalize())
    ax.legend()
    plt.show()

# Widget interactivo
interact(plot_polar_curve,
         curve_type=widgets.Dropdown(
             options=['butterfly', 'rose', 'cardioid'],
             value='butterfly',
             description='Curva:'
         ),
         a=widgets.FloatSlider(min=0.5, max=2.0, step=0.1, value=1, description='a'),
         b=widgets.FloatSlider(min=1, max=3, step=0.1, value=2, description='b'),
         c=widgets.IntSlider(min=5, max=20, step=1, value=12, description='c'),
         k=widgets.IntSlider(min=1, max=10, step=1, value=5, description='k')
);
```

##  Resultados  
Los experimentos numricos e interactivos nos permitieron observar lo siguiente:  

 **Curva Mariposa**: La forma de la curva cambia significativamente con los parmetros de la ecuacin, mostrando patrones simtricos y caticos dependiendo de los valores elegidos.  
 **Rosa Matemtica**: Se confirm que el nmero de ptalos de la rosa depende del parmetro \( k \), donde si \( k \) es impar, la curva tiene \( k \) ptalos, y si es par, tiene \( 2k \) ptalos.  
 **Cardioide**: Se verific que esta curva es un caso especial de la ecuacin de limacn y que su forma se asemeja a la de un corazn matemtico.  



##  Conclusin  
El estudio interactivo de las curvas polares nos permiti visualizar la belleza de las funciones trascendentes y comprender cmo los parmetros influyen en su forma geomtrica.  

 Se evidenci que la **Curva Mariposa** es un caso particular de curva polar altamente sensible a los cambios en su ecuacin.  
 La **Rosa Matemtica** y la **Cardioide** tambin mostraron patrones predecibles que pueden utilizarse en aplicaciones de diseo grfico, acstica y telecomunicaciones.
