# Propiedades de las funciones seno, coseno y tangente

## 游닀 Introducci칩n  
Las funciones trigonom칠tricas seno, coseno y tangente son fundamentales en el estudio de la geometr칤a y el an치lisis matem치tico. Estas funciones describen relaciones entre los 치ngulos y los lados de un tri치ngulo rect치ngulo y tienen aplicaciones en diversas 치reas como la f칤sica, la ingenier칤a y la computaci칩n.  

Para comprender mejor estas funciones, utilizamos el **c칤rculo unitario**, donde un 치ngulo **풪** se mide en sentido contrario a las agujas del reloj desde el eje **x positivo**. A partir de este c칤rculo, podemos derivar importantes propiedades que rigen el comportamiento del seno, coseno y tangente.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'


# Definir el 치ngulo 풪 en radianes (45춿 como ejemplo)
theta = np.pi / 4

# C칤rculo unitario
circle = plt.Circle((0, 0), 1, color='blue', fill=False, linestyle='dashed')

# Puntos clave
x = np.cos(theta)  # Coseno de 풪
y = np.sin(theta)  # Seno de 풪
tan_x = 1  # L칤nea tangente toca x = 1
tan_y = np.tan(theta)  # Valor de la tangente

# Crear la figura
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.set_xticks([-1, 0, 1])
ax.set_yticks([-1, 0, 1])
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Agregar el c칤rculo unitario
ax.add_patch(circle)

# Dibujar el radio del c칤rculo (hipotenusa)
ax.plot([0, x], [0, y], 'r-', linewidth=2, label=r'$r = 1$')

# Dibujar los catetos del tri치ngulo
ax.plot([x, x], [0, y], 'g--', label=r'$\sin(\theta)$')
ax.plot([0, x], [y, y], 'b--', label=r'$\cos(\theta)$')

# L칤nea tangente
ax.plot([1, 1], [0, tan_y], 'purple', linestyle='dotted', label=r'$\tan(\theta)$')

# Etiquetas de puntos clave
ax.text(x/2, y/2, r'$\theta$', fontsize=12, color='red', verticalalignment='bottom')
ax.text(x, -0.1, r'$\cos(\theta)$', fontsize=12, color='blue', horizontalalignment='center')
ax.text(-0.1, y, r'$\sin(\theta)$', fontsize=12, color='green', verticalalignment='center')
ax.text(1.05, tan_y/2, r'$\tan(\theta)$', fontsize=12, color='purple')

# Mostrar leyenda
ax.legend(loc='upper right')

# Mostrar la gr치fica
plt.title("游늷 C칤rculo Unitario y Funciones Trigonom칠tricas")
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

# Asegurar que Plotly renderiza en Google Colab
pio.renderers.default = "colab"

# Funci칩n para generar el gr치fico
def crear_grafico(theta_deg=45):
    theta = np.radians(theta_deg)
    x, y = np.cos(theta), np.sin(theta)
    tan_y = np.tan(theta) if np.abs(np.cos(theta)) > 1e-2 else None

    # Crear figura
    fig = go.Figure()

    # C칤rculo unitario
    t = np.linspace(0, 2*np.pi, 100)
    fig.add_trace(go.Scatter(x=np.cos(t), y=np.sin(t), mode="lines", name="C칤rculo Unitario"))

    # Tri치ngulo
    fig.add_trace(go.Scatter(x=[0, x, x], y=[0, 0, y], mode="lines", line=dict(color="red"), name="Tri치ngulo"))

    # Radio
    fig.add_trace(go.Scatter(x=[0, x], y=[0, y], mode="lines", line=dict(color="black", width=2), name="r = 1"))

    # Punto en el c칤rculo
    fig.add_trace(go.Scatter(x=[x], y=[y], mode="markers", marker=dict(color="black", size=8), name=f"P({x:.2f}, {y:.2f})"))

    # L칤neas seno y coseno
    fig.add_trace(go.Scatter(x=[x, x], y=[0, y], mode="lines", line=dict(color="green", dash="dash"), name="sen(풪)"))
    fig.add_trace(go.Scatter(x=[0, x], y=[y, y], mode="lines", line=dict(color="blue", dash="dash"), name="cos(풪)"))

    # L칤nea tangente
    if tan_y is not None and abs(tan_y) < 5:
        fig.add_trace(go.Scatter(x=[1, 1], y=[0, tan_y], mode="lines", line=dict(color="purple", dash="dot"), name="tan(풪)"))

    # Layout
    fig.update_layout(
        title=f"C칤rculo Unitario - 츼ngulo 풪 = {theta_deg}춿",
        xaxis=dict(range=[-1.2, 1.2], zeroline=True),
        yaxis=dict(range=[-1.2, 1.2], zeroline=True),
        showlegend=True,
        width=600, height=600
    )

    fig.show()  # Mostrar la gr치fica

# Llamar la funci칩n manualmente para ver el gr치fico inicial
crear_grafico(45)
```


## 游늵 Resultados  
Al representar un 치ngulo **풪** en el c칤rculo unitario:  
九 **El coseno (cos 풪)** corresponde a la coordenada **x** del punto sobre el c칤rculo.  
九 **El seno (sen 풪)** corresponde a la coordenada **y** del punto.  
九 **La tangente (tan 풪)** se define como la raz칩n entre el seno y el coseno, es decir, **$tan(풪) = sen(풪) / cos(풪)$**.  

### **Otras propiedades clave**  
游늷 **Relaciones fundamentales**  
- $( \sin^2(\theta) + \cos^2(\theta) = 1 ) $
- $( \tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)} )$  
- $( \tan(90^\circ) ) no\ est치\ definida\ porque ( \cos(90^\circ) = 0 )$.  

游늷 **Simetr칤as en el c칤rculo unitario**  
- $( \sin(-\theta) = -\sin(\theta) ) (funcion\ impar)$  
- $( \cos(-\theta) = \cos(\theta) ) (funcion\ par)$
- $( \tan(-\theta) = -\tan(\theta) ) $
