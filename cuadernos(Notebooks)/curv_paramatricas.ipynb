# Curvas ParamÃ©tricas

## ðŸ“– IntroducciÃ³n

Las **curvas paramÃ©tricas** son una forma de representar curvas en el espacio, donde las coordenadas `x` e `y` (o `z` en 3D) dependen de un parÃ¡metro comÃºn `t`.

ðŸ“Œ **Ejemplos comunes:**
- Circunferencia: x = cos(t), y = sin(t)
- Espiral: x = t cos(t), y = t sin(t)
- Lemniscata: x = cos(t) / (1 + sinÂ²(t)), y = cos(t) sin(t) / (1 + sinÂ²(t))

En este notebook, exploraremos curvas paramÃ©tricas en **2D y 3D**.

```python
# Importamos las bibliotecas necesarias
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt

### ðŸ”„ Definimos funciones para curvas paramÃ©tricas

def curva_circular(t):
    return np.cos(t), np.sin(t)

def curva_espiral(t):
    return t * np.cos(t), t * np.sin(t)

def curva_lemniscata(t):
    x = np.cos(t) / (1 + np.sin(t)**2)
    y = np.cos(t) * np.sin(t) / (1 + np.sin(t)**2)
    return x, y

### ðŸ“ˆ Graficamos curvas en 2D con Matplotlib

t = np.linspace(0, 2*np.pi, 400)
curvas = {"Circunferencia": curva_circular, "Espiral": curva_espiral, "Lemniscata": curva_lemniscata}

plt.figure(figsize=(10, 4))
for i, (nombre, funcion) in enumerate(curvas.items()):
    x, y = funcion(t)
    plt.subplot(1, 3, i+1)
    plt.plot(x, y, label=nombre)
    plt.title(nombre)
    plt.legend()
plt.show()

### ðŸ“Œ VisualizaciÃ³n en 3D con Plotly

def curva_parametrica_3d(t):
    return np.cos(t), np.sin(t), t / (2*np.pi)

t3d = np.linspace(0, 6*np.pi, 400)
x, y, z = curva_parametrica_3d(t3d)

fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(width=4, color='blue'))])
fig.update_layout(title="Curva ParamÃ©trica en 3D", scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```
