# **Comparaci√≥n entre $œÄ^e$ y $e^œÄ$**

## üìñ **Introducci√≥n**

Los n√∫meros irracionales $(\pi)$ y $(e)$ son fundamentales en matem√°ticas. En este estudio, analizaremos cu√°l de los valores $(\pi^e)$ o $(e^\pi)$ es mayor utilizando herramientas matem√°ticas como:

- Comparaci√≥n num√©rica directa.
- An√°lisis de funciones relacionadas.
- C√°lculo de m√°ximos y m√≠nimos mediante derivadas.

Este notebook proporciona una **explicaci√≥n visual e interactiva** para comprender mejor la relaci√≥n entre estas expresiones.


```python
# Importamos las bibliotecas necesarias
import numpy as np
import sympy as sp
import plotly.graph_objects as go

# Definimos los valores de œÄ y e
pi_val = np.pi
e_val = np.e

# Comparaci√≥n directa
pi_e = pi_val**e_val
e_pi = e_val**pi_val
resultado = "œÄ^e < e^œÄ" if pi_e < e_pi else "œÄ^e > e^œÄ"

# Definimos las funciones matem√°ticas para el an√°lisis
x = sp.symbols('x')
f = sp.exp(x) - 1 - x   # f(x) = e^x - 1 - x
g = sp.ln(x) / x        # g(x) = ln(x) / x
h = x**(1/x)            # h(x) = x^(1/x)

# Generamos valores para la gr√°fica
x_vals = np.linspace(0.1, 5, 100)

# Convertimos valores simb√≥licos a float
f_vals = [float(f.subs(x, val)) for val in x_vals]
g_vals = [float(g.subs(x, val)) for val in x_vals]
h_vals = [float(h.subs(x, val)) for val in x_vals]

# Puntos especiales en h(x) para x = e y x = œÄ
h_e = float(h.subs(x, e_val))
h_pi = float(h.subs(x, pi_val))

# Visualizaci√≥n interactiva con Plotly
fig = go.Figure()

# Funciones
fig.add_trace(go.Scatter(x=x_vals, y=f_vals, mode='lines', name='f(x) = e^x - 1 - x', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=x_vals, y=g_vals, mode='lines', name='g(x) = ln(x)/x', line=dict(color='green')))
fig.add_trace(go.Scatter(x=x_vals, y=h_vals, mode='lines', name='h(x) = x^(1/x)', line=dict(color='red', width=3)))

# Puntos clave en h(x)
fig.add_trace(go.Scatter(
    x=[e_val], y=[h_e], mode='markers', name='h(e)',
    marker=dict(color='red', size=10, symbol='circle')
))
fig.add_trace(go.Scatter(
    x=[pi_val], y=[h_pi], mode='markers', name='h(œÄ)',
    marker=dict(color='red', size=10, symbol='x')
))

# L√≠neas punteadas para destacar los valores
fig.add_trace(go.Scatter(x=[e_val, e_val], y=[0, h_e], mode='lines', line=dict(dash='dot', color='gray'), showlegend=False))
fig.add_trace(go.Scatter(x=[pi_val, pi_val], y=[0, h_pi], mode='lines', line=dict(dash='dot', color='gray'), showlegend=False))

# Configuraci√≥n del gr√°fico
fig.update_layout(
    title=f"üìä Comparaci√≥n de funciones para œÄ^e y e^œÄ: {resultado}",
    xaxis_title="x",
    yaxis_title="f(x), g(x), h(x)",
    legend_title="Funciones",
    margin=dict(l=20, r=20, t=40, b=20),
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True)
)

# Mostramos la gr√°fica interactiva
fig.show()
```

## üì¢ **Resultados**
La gr√°fica muestra el comportamiento de tres funciones matem√°ticas que se utilizan en demostraciones alternativas de la desigualdad:  

$$
\pi^e < e^\pi
$$

## Las funciones representadas son:  

### 1Ô∏è‚É£ $ f(x) = e^x - 1 - x)$  
- Representa la diferencia entre la exponencial y la recta ( y = x + 1 ).  
- Se usa para analizar convexidad y crecimiento exponencial.  

### 2Ô∏è‚É£ $( g(x) = {ln x}{x} )$  
- Se emplea para estudiar el m√°ximo de la funci√≥n \( x^{1/x} \), clave en la comparaci√≥n de potencias.  

### 3Ô∏è‚É£ $(h(x) = x^{1/x})$
- Permite visualizar en qu√© punto la base ( x ) elevada a su propio rec√≠proco alcanza su m√°ximo.  
- Se usa para comparar ( $\pi^e$ ) y $( e^\pi)$, ya que este m√°ximo ocurre en $( x = e )$.  

---

### üîé ¬øQu√© observamos en la gr√°fica?  

**La funci√≥n $( h(x) = x^{1/x})$ tiene un m√°ximo en $(x = e)$**  
- Esto significa que la base √≥ptima para maximizar potencias de la forma $( x^{1/x} ) es ( e )$.  
- Como $(\pi\ approx\ 3.14 )$ es mayor que $(e \approx 2.71 )$, se encuentra en la **parte decreciente** de $( h(x))$, lo que sugiere que:  
  
$$
pi^e < e^\pi
$$

**La funci√≥n $( g(x) = \ln x / x )$ tiene un m√°ximo en \( x = e \)**  
- Este comportamiento tambi√©n respalda que $( e^\pi )$ sea mayor que $( \pi^e )$.  

**La funci√≥n $( f(x) = e^x - 1 - x )$ crece r√°pidamente**  
- Indica que la funci√≥n exponencial supera linealmente a \( x + 1 \), lo cual es clave en ciertas demostraciones de la desigualdad.  


La gr√°fica confirma visualmente la desigualdad $( \pi^e < e^\pi )$, mostrando que la funci√≥n $( x^{1/x} )$ alcanza su m√°ximo en $( x = e )$, y que valores mayores a $( e )$ llevan a disminuciones en la funci√≥n, favoreciendo a $( e^\pi )$.  

üìå **Intuici√≥n matem√°tica:** La base √≥ptima para exponenciaciones de la forma $( x^{1/x} )$ es $( e )$, lo que explica por qu√© $( e^\pi )$ es mayor que $( \pi^e )$. üöÄ  






A trav√©s de este an√°lisis, obtenemos los valores:

- $( pi^e ) ‚âà {pi_e:.6f}$
- $( e^\pi ) ‚âà {e_pi:.6f}$

Observamos que **$( e^\pi)$** es mayor que **$( \pi^e )$** Este resultado puede demostrarse utilizando propiedades del c√°lculo y an√°lisis de funciones.
